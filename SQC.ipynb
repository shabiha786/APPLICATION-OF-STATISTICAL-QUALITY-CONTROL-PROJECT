```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load CSV file
import pandas as pd
df = pd.read_excel("/content/sample_data/smartphone.xlsx", engine="openpyxl")
print(df)
# Step 2: Calculate sample mean and range
xbar = df.mean(axis=1)
R = df.max(axis=1) - df.min(axis=1)

# Step 3: Calculate averages
xbar_bar = xbar.mean()
R_bar = R.mean()

# Step 4: Control chart constants (for sample size n=5)
A2 = 0.577
D3 = 0
D4 = 2.114

# Step 5: Control limits
UCL_x = xbar_bar + A2 * R_bar
LCL_x = xbar_bar - A2 * R_bar

UCL_R = D4 * R_bar
LCL_R = D3 * R_bar


```

```python
# Step 7: Plot R Chart
plt.figure(figsize=(8,5))
plt.plot(R, marker='o')
plt.axhline(R_bar, linestyle='--', label='CL')
plt.axhline(UCL_R, linestyle='--', label='UCL')
plt.axhline(LCL_R, linestyle='--', label='LCL')
plt.title("R Control Chart")
plt.xlabel("Sample")
plt.ylabel("Range")
plt.legend()
plt.show()
```

```python
# Step 6: Plot X-bar Chart
plt.figure(figsize=(8,5))
plt.plot(xbar, marker='o')
plt.axhline(xbar_bar, linestyle='--', label='CL')
plt.axhline(UCL_x, linestyle='--', label='UCL')
plt.axhline(LCL_x, linestyle='--', label='LCL')
plt.title("X-bar Control Chart")
plt.xlabel("Sample")
plt.ylabel("Sample Mean")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# number of defectives
d = np.array([5,6,4,7,5,6,8,5,4,6,7,5,9,6,4,5,7,6,8,5])

# sample size
n = 100

# proportion defective
p_bar = np.mean(d) / n

# control limits
CL = n * p_bar
UCL = n * p_bar + 3 * np.sqrt(n * p_bar * (1 - p_bar))
LCL = n * p_bar - 3 * np.sqrt(n * p_bar * (1 - p_bar))

# plot np chart
plt.plot(d, marker='o')
plt.axhline(CL, linestyle='--', label='CL')
plt.axhline(UCL, linestyle='--', label='UCL')
plt.axhline(LCL, linestyle='--', label='LCL')

plt.title("np Control Chart (EV Battery Overheating Defects)")
plt.xlabel("Sample Number")
plt.ylabel("Number of Defective Batteries")
plt.legend()

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Dataset
defects = [
    "Stitching Defect",
    "Sole Detachment",
    "Glue Marks",
    "Color Mismatch",
    "Size Variation",
    "Logo Misprint",
    "Material Tear"
]

frequency = [60,45,40,30,25,20,15]

df = pd.DataFrame({
    "Defect": defects,
    "Frequency": frequency
})

# Sort data
df = df.sort_values(by="Frequency", ascending=False)

# Cumulative percentage
df["Cum_Percent"] = df["Frequency"].cumsum() / df["Frequency"].sum() * 100

# Plot
fig, ax1 = plt.subplots(figsize=(10,6))

# Bar chart
ax1.bar(df["Defect"], df["Frequency"])
ax1.set_ylabel("Frequency")

# Rotate labels
plt.xticks(rotation=40, ha="right")

# Second axis for cumulative %
ax2 = ax1.twinx()
ax2.plot(df["Defect"], df["Cum_Percent"], marker="o")
ax2.set_ylabel("Cumulative % Defective")
ax2.set_ylim(0,120)

# Title
plt.title("Pareto Chart")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
issues = ["Overheating", "Cooling Failure", "Sensor Fault"]
cost = [26000, 18000, 9000]

plt.figure(figsize=(8,5))

bars = plt.bar(issues, cost, color=['red','orange','blue'])

# Add value labels
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 800, str(yval),
             ha='center', va='bottom', fontsize=11)

# Titles and labels
plt.title("Cost Impact of EV Battery Temperature Issues", fontsize=16)
plt.xlabel("Issue Type", fontsize=12)
plt.ylabel("Cost (₹)", fontsize=12)

# Increase Y-axis limit so top value shows properly
plt.ylim(0, 30000)

# Grid
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
